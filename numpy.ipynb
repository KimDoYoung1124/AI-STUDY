* 생성
  - 리스트 변환
  - 리스트 선언
  - 관련 : shape, size, dtype, ndim, reshape(형태변경), copy
---

```python
import numpy as np

# list 변환
list1 = [1, 2, 3]
a = np.array(list1)
print('a  = ', a)
print('a.shape = ', a.shape)
print('a.size  = ', a.size)
print('a.dtype = ', a.dtype)

# list 선언
b = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print('=============================')
print('b  = ', b)
print('b.shape = ', b.shape)
print('b.size  = ', b.size)
print('b.dtype = ', b.dtype)
```

* 초기화 관련 함수
  - zeros, ones, arange, reshape(형태 변경)
---

```python
# 0으로 초기화
np.zeros((3)) # , dtype='int64')
```

```python
# 1로 초기화
np.ones((3), dtype='int64')
```

```python
# arrange : arange(start, stop, step)
np.arange(0, 12, 2)
```

```python
# reshape 
a = np.arange(0, 12, 2).reshape(2, -1) # -1은 알아서
a
```

* 선택 : 슬라이싱, 정수 인덱싱, fancy indexing
---

```python
# 슬라이싱
lst = [[1, 2, 3],
       [4, 5, 6],
       [7, 8, 9]]
arr = np.array(lst)
a = arr[0:2, 0:2] # 슬라이스
print(a)
```

```python
# 정수 인덱싱
arr[[0,1],[1,0]] # 0행 1열, 1행 0열
```

```python
# fancy indexing TRUE, FALSE로
a_5 = b > 5
print(a_5)
```
