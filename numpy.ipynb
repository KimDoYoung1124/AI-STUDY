2. 리스트 선언

```python
import numpy as np

# list 변환
list1 = [1, 2, 3]
a = np.array(list1)
print('a  = ', a)
print('a.shape = ', a.shape)
print('a.size  = ', a.size)
print('a.dtype = ', a.dtype)

# list 선언
b = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print('=============================')
print('b  = ', b)
print('b.shape = ', b.shape)
print('b.size  = ', b.size)
print('b.dtype = ', b.dtype)
```
